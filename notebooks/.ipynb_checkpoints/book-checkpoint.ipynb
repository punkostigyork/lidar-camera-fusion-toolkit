# KITTI Sensor Fusion: Multi-Sensor 3D Perception
This toolkit provides a pipeline for fusing LiDAR point clouds with camera imagery and visualizing 3D Ground Truth labels.

### Pipeline Stages:
1. **Environment Setup:** Path configuration and custom module imports.
2. **Point Cloud Fusion:** Projecting raw LiDAR points onto the 2D image plane.
3. **3D Object Visualization:** Projecting 3D tracklets and wireframes.
4. **Bird's Eye View (BEV):** Generating a metric top-down map with distance grids.
5. **Sequence Processing:** Automated video generation.

```python
import sys
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Add project root to path
root_path = os.path.abspath(os.path.join('..'))
if root_path not in sys.path:
    sys.path.append(root_path)

%load_ext autoreload
%autoreload 2

from src.core.projection import Projector
from src.loaders.kitti_loader import KittiLoader, KittiCalib
from src.loaders.kitti_labels import KittiLabels
from src.core.sequence_processor import SequenceProcessor

print("✅ Setup Complete.")
```

## 1. Initialization
Configure data paths and initialize the calibration-aware Projector.

```python
DATA_PATH = "../data/kitti_sample"
loader = KittiLoader(DATA_PATH)
calib = KittiCalib(f"{DATA_PATH}/calib")
labels = KittiLabels(f"{DATA_PATH}/tracklet_labels.xml")
projector = Projector(calib)

os.makedirs("../outputs", exist_ok=True)
print("✅ All data components loaded.")
```

## 2. Verification: LiDAR-Camera Fusion
Testing the extrinsic calibration by projecting raw LiDAR points into the camera image.

```python
img = loader.load_image(0)
pc = loader.load_lidar(0)

# Project points
pixels, depths = projector.project_to_image(pc)

# Filter for image bounds and depth range
h, w, _ = img.shape
mask = (pixels[:, 0] >= 0) & (pixels[:, 0] < w) & \
       (pixels[:, 1] >= 0) & (pixels[:, 1] < h) & \
       (depths > 2) & (depths < 40)

valid_pixels = pixels[mask].astype(int)
valid_depths = depths[mask]

for i in range(len(valid_pixels)):
    x, y = valid_pixels[i]
    color_val = int(min(valid_depths[i], 30) / 30 * 255)
    cv2.circle(img, (x, y), 2, (255 - color_val, color_val, 255), -1)

plt.figure(figsize=(15, 5))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("LiDAR Points Projected to Image Plane")
plt.axis('off')
plt.show()
```

## 3. Verification: 3D Ground Truth Boxes
Verifying the tracklet label parsing and 3D coordinate transformation.

```python
img = loader.load_image(0)
boxes = labels.get_boxes_for_frame(0)

for box in boxes:
    corners_3d = projector.get_3d_box_corners(box)
    corners_2d, depths = projector.project_to_image(corners_3d)
    
    if np.all(depths > 0):
        color = (0, 255, 0) if box['type'] == 'Car' else (255, 0, 0)
        img = projector.draw_3d_box(img, corners_2d, color=color)

plt.figure(figsize=(15, 5))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Ground Truth 3D Bounding Boxes")
plt.axis('off')
plt.show()
```

## 4. Final Perception Dashboard
Combining all features into a side-by-side Camera and Bird's Eye View visualization.

```python
frame_idx = 0
img = loader.load_image(frame_idx)
pc = loader.load_lidar(frame_idx)
boxes = labels.get_boxes_for_frame(frame_idx)

for box in boxes:
    corners_3d = projector.get_3d_box_corners(box)
    corners_2d, depths = projector.project_to_image(corners_3d)
    if np.all(depths > 0):
        color = (0, 255, 0) if box['type'] == 'Car' else (255, 255, 0)
        img = projector.draw_3d_box(img, corners_2d, color=color)
        img = projector.draw_label(img, box, corners_2d)

bev_map = projector.generate_bev(boxes, pc=pc, width=400, height=600, scale=10)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(22, 10), gridspec_kw={'width_ratios': [3, 1]})
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title("Front Camera View")
ax1.axis('off')

ax2.imshow(cv2.cvtColor(bev_map, cv2.COLOR_BGR2RGB))
ax2.set_title("Bird's Eye View (Metric Map)")
ax2.axis('off')

plt.tight_layout()
plt.show()
```

## 5. Sequence Export
Exporting the final fused visualization to a high-definition video.

```python
processor = SequenceProcessor(loader, projector, labels)
processor.process_sequence("../outputs/final_perception_video.mp4", num_frames=154)
print("🚀 Processing complete. Video saved to outputs/.")
```
