# KITTI Sensor Fusion: 3D Object Detection & Bird's Eye View
This notebook demonstrates a complete sensor fusion pipeline using the KITTI dataset. 

### Key Features:
* **3D Projection:** Projecting 3D bounding boxes from LiDAR/Camera coordinates to 2D image pixels.
* **Bird's Eye View (BEV):** Generating a metric top-down map of the environment.
* **Distance Estimation:** Real-time distance calculation for tracked objects.
* **LiDAR Fusion:** Overlaying point cloud data onto camera images.

```python
import sys
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 1. Setup Project Paths
root_path = os.path.abspath(os.path.join('..'))
if root_path not in sys.path:
    sys.path.append(root_path)

# 2. Enable Autoreload
%load_ext autoreload
%autoreload 2

from src.core.projection import Projector
from src.loaders.kitti_loader import KittiLoader, KittiCalib
from src.loaders.kitti_labels import KittiLabels
from src.core.sequence_processor import SequenceProcessor

print("✅ Environment Setup Complete.")
```

## 1. Data Initialization
We initialize the loaders for Calibration, Labels (XML), and the raw Data (Images/LiDAR).

```python
DATA_PATH = "../data/kitti_sample"
LABEL_PATH = f"{DATA_PATH}/tracklet_labels.xml"
CALIB_PATH = f"{DATA_PATH}/calib"

loader = KittiLoader(DATA_PATH)
calib = KittiCalib(CALIB_PATH)
labels = KittiLabels(LABEL_PATH)
projector = Projector(calib)

os.makedirs("../outputs", exist_ok=True)
print("✅ Data components initialized.")
```

## 2. Multi-View Verification
This cell generates a side-by-side dashboard to verify spatial alignment between the Camera and the BEV Map.

```python
# Select frame for inspection
frame_idx = 0
img = loader.load_image(frame_idx)
pc = loader.load_lidar(frame_idx)
boxes = labels.get_boxes_for_frame(frame_idx)

# Process Camera View
for box in boxes:
    corners_3d = projector.get_3d_box_corners(box)
    corners_2d, depths = projector.project_to_image(corners_3d)
    
    if np.all(depths > 0):
        color = (0, 255, 0) if box['type'] == 'Car' else (255, 255, 0)
        img = projector.draw_3d_box(img, corners_2d, color=color)
        img = projector.draw_label(img, box, corners_2d)

# Process Bird's Eye View (BEV)
bev_map = projector.generate_bev(boxes, pc=pc, width=400, height=600, scale=10)

# Display Results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(22, 10), gridspec_kw={'width_ratios': [3, 1]})
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title(f"Front Camera View - Frame {frame_idx}")
ax1.axis('off')

ax2.imshow(cv2.cvtColor(bev_map, cv2.COLOR_BGR2RGB))
ax2.set_title("Bird's Eye View (Metric Map)")
ax2.axis('off')

plt.tight_layout()
plt.show()
```

## 3. Full Sequence Processing
Generate the final MP4 video export with all overlays enabled.

```python
processor = SequenceProcessor(loader, projector, labels)
output_file = "../outputs/kitti_final_dashboard.mp4"

print("Starting sequence processing...")
processor.process_sequence(output_file, num_frames=154)
print(f"🚀 Video saved successfully to: {output_file}")
```
